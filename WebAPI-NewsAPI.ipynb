```python
# Import dependencies

import requests
import time
```

```python
# Settiing up NEWS API
# Query for mental health services between January 01, 2020 to August 25, 2020

url = "http://newsapi.org/v2/everything?"
api_key= "YOUR API KEY HERE"

# Store a search term
query = "mental health services, COVID"

# Search for articles published between a begin and end date
begin_date = "20200101"
end_date = "20200825"

query_url = f"{url}q={query}&from_param={begin_date}&to={end_date}&apiKey={api_key}"
```

```python
# Retrieve articles - First 10
articles = requests.get(query_url).json()
articles_list = articles["articles"]

for article in articles_list:
    print(article["publishedAt"])
    print(article["title"])
    print(f'A snippet from the article: {article["description"]}')
    print(article["url"])
    print('---------------------------')
```
